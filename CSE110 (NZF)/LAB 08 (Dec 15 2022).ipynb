```python
# Task 01

def bubbleSortAscending(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-1-i):
            if lst[j+1] < lst[j]:
                temp = lst[j] 
                lst[j] = lst[j+1]
                lst[j+1] = temp
    return lst

list1 = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

print(bubbleSortAscending(list1))
```

```python
# Task 02

def selectionSortAscending(lst):
    for curr_i in range(len(lst)):
        min_i = curr_i
        for j in range(curr_i+1, len(lst)):
            if lst[j] < lst[min_i]:
                min_i = j
        temp = lst[curr_i] 
        lst[curr_i] = lst[min_i]
        lst[min_i] = temp
    return lst

my_list= [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

print(selectionSortAscending(my_list))
```

```python
# Task 03

def bubbleSortDescending(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-1-i):
            if lst[j+1] > lst[j]:
                temp = lst[j] 
                lst[j] = lst[j+1]
                lst[j+1] = temp
    return lst

list1 = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

print(bubbleSortDescending(list1))
```

```python
# Task 05

def sort_students(course, students):
  # Create a dictionary with student names as keys and marks as values
  student_marks = {}
  for student in students:
    name = student[0]
    # Get the index of the course
    if course == "CSE110":
      index = 1
    elif course == "PHY111":
      index = 2
    elif course == "MAT110":
      index = 3
    # Add the student name and mark to the dictionary
    student_marks[name] = student[index]
  # Sort the dictionary in descending order based on the marks
  sorted_students = sorted(student_marks, key=student_marks.get, reverse=True)
  return sorted_students

# List of students
students = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]

# Take the course name as input from the user
course = input("Enter the course name: ")

# Call the function and print the sorted list of students
sorted_students = sort_students(course, students)
print(sorted_students)
```

```python
# Task 08

def find_closest_pairs(numbers):
  # Sort the list in ascending order
  numbers.sort()

  # Initialize variables to store the pairs with the smallest sum
  smallest_sum = float("inf")
  closest_pair = (0, 0)

  # Iterate through the list and find the pairs with the smallest sum
  for i in range(len(numbers) - 1):
    if abs(numbers[i] + numbers[i+1]) < smallest_sum:
      smallest_sum = abs(numbers[i] + numbers[i+1])
      closest_pair = (numbers[i], numbers[i+1])

  return "Two pairs which have the smallest sum = {} and {}".format(closest_pair[0], closest_pair[1])

# Test the function
list_one = [-10, 15, 2, 4, -4, 7, -8]
print(find_closest_pairs(list_one))

list_two = [1, -8, 4, -7, -20, 26, 70, -85]
print(find_closest_pairs(list_two))
```
