```python
# Task 02

def fibonacci(num):
    f_value = 0
    s_value = 1

    for i in range(num + 1):
        if f_value > num:
            break
            
        else:
            print(f_value, end = " ")
            next = f_value + s_value
            f_value = s_value
            s_value = next

fibonacci(10)
```

```python
# Task 04

def upperlower_counter(string):
    upper_count = 0
    lower_count = 0

    for i in string:
        if 65 <= ord(i) <= 90:
            upper_count += 1
        elif 97 <= ord(i) <= 122:
            lower_count += 1

    print("No of Uppercase characters: ", upper_count)
    print("No of Lowercase characters: ", lower_count)

upperlower_counter("The quick Sand Man")
```

```python
# Task 08

def show_palindrome(num):
    for i in range(1, num+1):
        print(i, end = " ")
    for j in range(num-1, 0, -1):
        print(j, end = " ")
    return ""

def show_palindrome_triangle(n):
    for i in range(1, n+1):
        print("  " * (n-i), end = "")
        print(str(show_palindrome(i)))
        
show_palindrome_triangle(5)
```

```python
# Task 11

def rem_duplicate(tup):
    t = list(tup)
    l = []
    
    for i in t:
        if i not in l:
            l.append(i)
    return tuple(l)

c = rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2]))
print(c)
```

```python
# Bonus Task 02
# Similar to 8th task of this lab assignment, you just have to create a pyramid,
# not once, but twice! Thus, make a DIAMOND. So yeah, there're few simple steps now:

def diamond_pattern(n):
  # Create a list of numbers from 1 to n
  numbers = [str(i) for i in range(1, n+1)]

  # Create the top half of the diamond
  for i in range(n):
    print(" " * (n-i-1) + " ".join(numbers[:i+1]) + " " * (n-i-1))

  # Create the bottom half of the diamond
  for i in range(n-2, -1, -1):
    print(" " * (n-i-1) + " ".join(numbers[:i+1]) + " " * (n-i-1))

# Test the function with n = 5
diamond_pattern(5)

# https://youtu.be/ybgdlPLwZH8
```
