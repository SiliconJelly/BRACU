```python
# Task 08

key_and_value = input().split(", ")
dict1 = {}

for i in key_and_value:
    temp = ""
    c = 0
    for j in i:
        if j == ":":
            dict1.update({temp:i[c+1::]})
            temp = ""
            break
        else:
            temp += j
            c += 1
count = 0
sum = 0

for k in dict1:
    sum += int(dict1[k])
    count += 1

# ðŸ‘‡ï¸ {'Jon': '100', 'Dan': '200', 'Rob': '300'}
print("Average is", int(sum/count))
```

```python
# Task 10

given_dict = {'sci fi': 12,'mystery': 15,'horror': 8,'mythology': 10,'young_adult': 4,'adventure':14}
list1 = []

for i in given_dict:
    value = given_dict[i]
    list1.append(value)

max = list1[0]

for j in list1:
    if j > max:
        max = j
for k in given_dict:
    if given_dict[k] == max:
        key = k

print(f"The highest selling book genre is '{key}' and the number of books sold are {max}.")
```

```python
# Task 11

inp = input()
dup_inp = ''
for h in inp:
    dup_inp+=h.lower()

key = []
d = {}
c = 0
for i in dup_inp:
    if i == '"' or i==" ":
        continue
    elif i not in key:
        key.append(i)
for i in key:
    for j in dup_inp:
        if j == i:
            c+=1
    d.update({i:c})
    c = 0
    
print(d)
```

```python
# Task 13

list_1 = [("a", 1),("b", 2),("a", 3),("b", 1),("a", 2),("c", 1)]
key=[]
values=[]
conf_value=[]
d={}

for i in list_1:
    if i[0] not in key:
        key.append(i[0])
        for j in list_1:
            if i[0] in j:
                values.append(j[1])
        for h in values:
            if h not in conf_value:
                conf_value.append(h)
        d.update({i[0]:conf_value})
        values=[]
        conf_value=[]
        
print(d)
```
