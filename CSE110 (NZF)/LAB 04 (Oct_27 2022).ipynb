```python
# string input will have alphabets and spaces
# checks if "small letters" are gradually increasing
# unique small letters are printed separately, and
# another string with ascii values get printed as well

small = ""  # stores the small alphabet
ascii = ""  # concatenates ascii value
ans = ""    # removes repeated alphabet
flag = True

S = input()

for e in S:
  if ord("a") <= ord(e) <= ord("z"):
    small += e

for i in range(len(small)-1):

  if ord(small[i]) < ord(small[i+1]): # increment / sequence checking
    ans += small[i]
    ascii += str(ord(small[i]))

  elif ord(small[i]) == ord(small[i+1]):
    continue # ignores the repeated alphabet

  else:
    print("Not increasing")
    flag = False
    break


if flag != False:
  print(ans)
  print(ascii)
```
