```python
# Task 22

canvas_price = 120
paint_price = 75
discount = 0

canvas_amount = int(input())
paint_amount = int(input())

total = (canvas_price * canvas_amount) + (paint_price * paint_amount)

if total >= 0 and total <= 299:
# if 0 >= total <= 299:
  discount = 0

elif 300 >= total <= 499:
  discount = 10

elif 500 >= total <= 749:
  discount = 20

elif 750 >= total <= 999:
  discount = 50

else:
  discount = 150


print("Previous total:", total)
print("New total after discount:", total - discount)
```
