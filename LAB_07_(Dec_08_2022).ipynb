```python
# Task 02

def fibonacii(num):
    f_value=0
    s_value=1

    for i in range(num+1):
        if f_value>num:
            break
            
        else:
            print(f_value,end=" ")
            next=f_value+s_value
            f_value=s_value
            s_value=next

fibonacii(10)
```

```python
# Task 04

def upperlower_counter(string):
    upper_count=0
    lower_count=0

    for i in string:
        if 65<=ord(i)<=90:
            upper_count+=1
        elif 97<=ord(i)<=122:
            lower_count+=1

    print("No of Uppercase characters: ",upper_count)
    print("No of Lowercase characters: ",lower_count)

upperlower_counter("The quick Sand Man")
```

```python
#Task 08

def show_palindrom(num):
    for i in range(1,num+1):
        print(i,end=" ")
    for j in range(num-1,0,-1):
        print(j,end=" ")
    return ""

def show_palindrom_triangle(n):
    for i in range(1,n+1):
        print("  "*(n-i),end="")
        print(str(show_palindrom(i)))
        
show_palindrom_triangle(5)
```

```python
# Task 11

def rem_duplicate(tup):
    t=list(tup)
    l=[]
    
    for i in t:
        if i not in l:
            l.append(i)
    return tuple(l)

c=rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2]))
print(c)
```

```python
# Task 13

def calculator(op,n1,n2):
    if op=="+":
        print(float(n1+n2))
    elif op=="-":
        print(float(n1-n2))
    elif op=="*":
        print(float(n1*n2))
    else:
        print(n1/n2)

op=input()
n1=float(input())
n2=float(input())
calculator(op,n1,n2)
```

```python
# Task 16

n = int(input())

def splitting_money(n):
    s=""
    while True:
        if n>0:
            if n>=500:
                note=n//500
                a="500 Taka: "+str(note)+" note(s)"
                n=n%500
            elif 500>n>=100:
                note=n//100
                a="100 Taka: "+str(note)+" note(s)"
                n=n%100
            elif 100>n>=50:
                note=n//50
                a="50 Taka: "+str(note)+" note(s)"
                n=n%50
            elif 50>n>=20:
                note=n//20
                a="20 Taka: "+str(note)+" note(s)"
                n=n%20
            elif 20>n>=10:
                note=n//10
                a="10 Taka: "+str(note)+" note(s)"
                n=n%10
            elif 10>n>=5:
                note=n//5
                a="5 Taka: "+str(note)+" note(s)"
                n=n%5
            elif 5>n>=2:
                note=n//2
                a="2 Taka: "+str(note)+" note(s)"
                n=n%2
            else:
                note=n//1
                a="1 Taka: "+str(note)+" note(s)"
                n=n%1
            s+=f"{a}\n"
        else:
            break
    return s

sm=splitting_money(n)
print(sm)
```

```python
# Task 17

def remove_odd(l):
    r_l=[]
    for i in l:
        if i%2==0:
            r_l.append(i)
    return r_l
    
removed_odd_list=remove_odd([11,2,3,4,5,2,0,5,3])
print(removed_odd_list)
```

```python
# Task 18

start=int(input())
end=int(input())
first_div=int(input())
second_div=int(input())

def divisible_total(a,b,c,d):
    total=0
    for i in range(a,b):
        if i%c==0 and i%d==0:
            pass
        elif i%c==0 or i%d==0:
            total+=i
    return total

total=divisible_total(start,end,first_div,second_div)
print(total)
```
